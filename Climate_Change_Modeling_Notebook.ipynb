```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_style('whitegrid')
```

```python
# Step 2: Load dataset
url = r"C:\Users\MY HP\OneDrive\Desktop\archive.csv"
df = pd.read_csv(url)

# Preview the first 5 rows
df.head()
```

```python
# Step 3: Check for missing values
df.isnull().sum()
```

```python
# Step 4: Summary statistics of the dataset
df.describe()
```

```python
# Step 5: Line plot for CO2 trends
plt.figure(figsize=(14,6))
plt.plot(df['Decimal Date'], df['Carbon Dioxide (ppm)'], label='CO2 ppm', color='green')
plt.plot(df['Decimal Date'], df['Seasonally Adjusted CO2 (ppm)'], label='Seasonally Adjusted CO2', color='red')
plt.xlabel('Year')
plt.ylabel('CO2 Concentration (ppm)')
plt.title('CO2 Levels Over Time')
plt.legend()
plt.show()
```

```python
# Step 6: Plot CO2 fit trends
plt.figure(figsize=(14,6))
plt.plot(df['Decimal Date'], df['Carbon Dioxide Fit (ppm)'], label='CO2 Fit', linestyle='--')
plt.plot(df['Decimal Date'], df['Seasonally Adjusted CO2 Fit (ppm)'], label='Seasonally Adjusted CO2 Fit', linestyle=':')
plt.xlabel('Year')
plt.ylabel('CO2 Concentration (ppm)')
plt.title('CO2 Fit Trends')
plt.legend()
plt.show()
```

```python
# Step 7: Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
